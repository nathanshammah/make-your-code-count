```python
import numpy as np
from scipy import *
import matplotlib.pyplot as plt
```

```python
# Create a vector from 0 to 10 with 100 elements
vector = np.linspace(0,10,100)
function_sin = np.sin(vector) 
```

```python
# Plot a figure
plt.figure()
plt.plot(vector, function_sin)
plt.show()
plt.close()
```
